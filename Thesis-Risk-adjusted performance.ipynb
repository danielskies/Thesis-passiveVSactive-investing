```python
import pandas as pd

# First, let's isolate the relevant data
etf_df = daily_returns[ticker_etf].copy()
mf_df = daily_returns[ticker_mf].copy()

# Then, create equally weighted portfolio returns
etf_portfolio_returns = etf_df.mean(axis=1)
mf_portfolio_returns = mf_df.mean(axis=1)
etf_portfolio_returns = etf_portfolio_returns.to_frame(name='Returns')
mf_portfolio_returns = mf_portfolio_returns.to_frame(name='Returns')

etf_portfolio_returns
```

```python
import yfinance as yf

# Download S&P 500 data (only Adj Close) from 2016 to present
ticker = '^SP500TR'  # Ticker symbol for S&P 500
start_date = '2016-01-01'
end_date = '2023-04-29'
sp500_adjclose = yf.download(ticker, start=start_date, end=end_date)['Adj Close']
sp500_returns = sp500_adjclose.pct_change().dropna()
sp500_returns
```

```python
etf_portfolio = pd.merge(etf_portfolio_returns, rf_daily, left_index=True, right_index=True)
mf_portfolio = pd.merge(mf_portfolio_returns, rf_daily, left_index=True, right_index=True)
sp500 = pd.merge(sp500_returns, rf_daily, left_index=True, right_index=True)
sp500 = sp500.rename(columns={'Adj Close': 'Returns'})

sp500
```

```python
mf_portfolio['ExcessReturns'] = mf_portfolio['Returns'] - mf_portfolio['rf']
etf_portfolio['ExcessReturns'] = etf_portfolio['Returns'] - etf_portfolio['rf']
sp500['ExcessReturns'] = sp500['Returns'] - sp500['rf']
mf_portfolio
```

```python
mf_portfolio['Volatility'] = mf_portfolio['Returns'].rolling(window=30, min_periods=1).std()
etf_portfolio['Volatility'] = etf_portfolio['Returns'].rolling(window=30, min_periods=1).std()
sp500['Volatility'] = sp500['Returns'].rolling(window=30, min_periods=1).std()

mf_portfolio
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

# Define the rolling window size
window_size = 3

# Create empty lists to store alphas and betas
alphas = []
betas = []

# Iterate over the DataFrame with a rolling window
for i in range(len(mf_portfolio) - window_size + 1):
    # Get the window of excess returns and benchmark excess returns
    window_excess_returns = mf_portfolio['ExcessReturns'].iloc[i : i + window_size]
    window_benchmark_excess_returns = sp500['ExcessReturns'].iloc[i : i + window_size]

    # Perform linear regression
    X = sm.add_constant(window_benchmark_excess_returns)
    model = sm.OLS(window_excess_returns, X)
    results = model.fit()

    # Extract alpha and beta
    alpha = results.params[0]
    beta = results.params[1]

    # Append alpha and beta to the respective lists
    alphas.append(alpha)
    betas.append(beta)

# Pad the beginning of the lists with NaN values to match the DataFrame length
alphas = [np.nan] * (window_size - 1) + alphas
betas = [np.nan] * (window_size - 1) + betas

# Add alphas and betas to DataFrame
mf_portfolio['Alpha'] = alphas
mf_portfolio['Beta'] = betas
mf_portfolio
```

```python
import numpy as np

# Calculate the difference between portfolio returns and benchmark returns
tracking_error_etf = etf_portfolio['Returns'] - sp500['Returns']
tracking_error_volatility_etf = tracking_error_etf.rolling(window=30, min_periods=1).std()
tracking_error_mf = mf_portfolio['Returns'] - sp500['Returns']
tracking_error_volatility_mf = tracking_error_mf.rolling(window=30, min_periods=1).std()

mf_portfolio['TrErr'] = tracking_error_volatility_mf
etf_portfolio['TrErr'] = tracking_error_volatility_etf
etf_portfolio
```

```python
import numpy as np

# Convert daily values to annualized values
trading_days_per_year = 252  

# Beta does not need to be annualized

mean_to_annualize = ['Returns', 'rf','ExcessReturns','Alpha']
mf_portfolio[mean_to_annualize] = mf_portfolio[mean_to_annualize] * trading_days_per_year

# Apply annualization to specific columns
std_to_annualize = ['Volatility', 'TrErr']
mf_portfolio[std_to_annualize] = mf_portfolio[std_to_annualize] * np.sqrt(trading_days_per_year)
```
