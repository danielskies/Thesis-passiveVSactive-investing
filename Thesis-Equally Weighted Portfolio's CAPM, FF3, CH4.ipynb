```python
# Create empty lists to store the results
stocks = []
alphas = []
betas = []
smb_coefs = []
hml_coefs = []
RSQ = []
alpha_pvalues = []

# Iterate over each stock
for stock in normal_returns_df.columns:
    # Remove rows with NaN values from both the stock and index returns
    combined_returns = pd.concat([normal_sp500_excess_returns, normal_returns_df[stock], SMB_normal, HML_normal], axis=1).dropna()
    
    # Check if the combined_returns DataFrame is empty
    if combined_returns.empty:
        # Handle empty DataFrame case (e.g., by assigning NaN values to coefficients)
        alpha, beta, smb_coef, hml_coef, constant = (float('NaN'),) * 5
        alpha_pvalue = float('NaN')
    else:
        # Perform linear regression to calculate coefficients
        Y = combined_returns.iloc[:, 1]
        X = combined_returns.iloc[:, [0, 2, 3]]
        X = sm.add_constant(X)
        model = sm.OLS(Y, X)
        results = model.fit()
        
        # Assign coefficients and p-value
        params = results.params
        if len(params) >= 4:
            alpha, beta, smb_coef, hml_coef = params[:4]
            constant = params[4] if len(params) == 5 else float('NaN')
        else:
            alpha, beta, smb_coef, hml_coef, constant = (float('NaN'),) * 5
        alpha_pvalue = results.pvalues[0]
        
    # Append the results to the lists
    stocks.append(stock)
    alphas.append(alpha)
    betas.append(beta)
    smb_coefs.append(smb_coef)
    hml_coefs.append(hml_coef)
    RSQ.append(results.rsquared if not combined_returns.empty else float('NaN'))
    alpha_pvalues.append(alpha_pvalue)

# Create a DataFrame from the results
FF3_20 = pd.DataFrame({'Fund': stocks, 'Alpha': alphas, 'Market Beta': betas, 'SMB Coef': smb_coefs, 'HML Coef': hml_coefs,
                      'R-Squared': RSQ, 'p-value a': alpha_pvalues})

# Round the p-values to a reasonable number of decimal places (e.g., 5)
FF3_20['p-value a'] = FF3_20['p-value a'].round(5)
# Print the results
FF3_20
```
