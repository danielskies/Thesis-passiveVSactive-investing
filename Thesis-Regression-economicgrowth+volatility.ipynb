```python
# Assign dates
start_date_normal = '2016-01-04'
end_date_normal = '2019-12-31'
start_date_crisis = '2020-01-01'
end_date_crisis = '2023-04-28'
```

```python
import pandas_datareader as pdr
import numpy as np

# Define the Treasury rate series code
treasury_rate_series = 'DGS1'

# Download Treasury rates
treasury_rates = pdr.get_data_fred(treasury_rate_series, start_date_normal, end_date_crisis)

# Print the downloaded data
rf_yearly = treasury_rates/100
rf_yearly = rf_yearly.dropna()
rf_daily = np.exp(rf_yearly / 252) - 1
rf_daily.index.names = ['Date']  # Rename the index of etf_portfolio_returns
rf_daily = rf_daily.rename(columns={'DGS1': 'rf'})
rf_daily
```

```python
daily_returns = adj_close_stocks_filtered.pct_change()
daily_returns = daily_returns.iloc[1:]
```

```python
import pandas as pd

# First, let's isolate the relevant data
etf_df = daily_returns[ticker_etf].copy()
mf_df = daily_returns[ticker_mf].copy()

# Then, create equally weighted portfolio returns
etf_portfolio_returns = etf_df.mean(axis=1)
mf_portfolio_returns = mf_df.mean(axis=1)
etf_portfolio_returns = etf_portfolio_returns.to_frame(name='Returns')
mf_portfolio_returns = mf_portfolio_returns.to_frame(name='Returns')
```

```python
import yfinance as yf

# Download S&P 500 data (only Adj Close) from 2016 to present
ticker = '^SP500TR'  # Ticker symbol for S&P 500
start_date = '2016-01-01'
end_date = '2023-04-29'
sp500_adjclose = yf.download(ticker, start=start_date, end=end_date)['Adj Close']
sp500_returns = sp500_adjclose.pct_change().dropna()
```

```python
etf_portfolio = pd.merge(etf_portfolio_returns, rf_daily, left_index=True, right_index=True)
mf_portfolio = pd.merge(mf_portfolio_returns, rf_daily, left_index=True, right_index=True)
sp500 = pd.merge(sp500_returns, rf_daily, left_index=True, right_index=True)
sp500 = sp500.rename(columns={'Adj Close': 'Returns'})
```

```python
mf_portfolio['ExcessReturns'] = mf_portfolio['Returns'] - mf_portfolio['rf']
etf_portfolio['ExcessReturns'] = etf_portfolio['Returns'] - etf_portfolio['rf']
sp500['ExcessReturns'] = sp500['Returns'] - sp500['rf']
```

```python
mf_portfolio['Volatility'] = mf_portfolio['Returns'].rolling(window=30, min_periods=1).std()
etf_portfolio['Volatility'] = etf_portfolio['Returns'].rolling(window=30, min_periods=1).std()
sp500['Volatility'] = sp500['Returns'].rolling(window=30, min_periods=1).std()
```

```python
sp500_vol = sp500_returns.rolling(window=30, min_periods=1).std()
sp500_vol = sp500_vol.drop(sp500_vol.index[0])
```

```python
import pandas_datareader as pdr

# Define the GDP series code
gdp_series = 'GDP'

# Define your start and end dates
start_date = '2016-01-01'
end_date = '2023-05-31'

# Download GDP data
gdp = pdr.get_data_fred(gdp_series, start_date, end_date)

# Convert to daily frequency using linear interpolation
gdp_daily = gdp.resample('D').interpolate('linear')
gdp_daily
```

```python
import pandas_datareader.data as web
import datetime

# Define start and end dates
start = datetime.datetime(2016, 1, 1)
end = datetime.datetime(2023, 12, 31)

# FRED codes for 1-month, 1-year, and 10-year Treasury yield
series = ['DGS1MO', 'DGS1', 'DGS10']

# Fetch the data
yield_data = web.DataReader(series, 'fred', start, end)

# Compute the slope of the yield curve
yield_data['YieldCurve'] = yield_data['DGS10'] - yield_data['DGS1MO']

# Compute the daily percentage change of the yield curve
yield_data['YieldCurve_pct_change'] = yield_data['YieldCurve'].pct_change()

# Drop missing values
yield_data = yield_data.dropna()

# Print the data
yield_data
```
